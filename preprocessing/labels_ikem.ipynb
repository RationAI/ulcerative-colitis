```python
from pathlib import Path

import pandas as pd


DIR = Path("/mnt/data/Projects/Data/IKEM/colon/IBD_AI/20x")
```

```python
# Cohort from Dec 2024

df1 = pd.read_csv(DIR / "Fab_IBD_AI_12_2024.csv", index_col=0)
df1.index = df1.index.str.lstrip("0")
df1["Lokalita"] = df1["Lokalita"].replace(
    {"cekoascendens": "cekum", "rektosigma": "rektum"}
)
df1 = df1.fillna("rektum")
df1
```

```python
# Cohort from May 2025

df2 = pd.read_excel(DIR / "IBD_AI_2.xlsx", index_col=0)
df2 = df2.rename(columns={"lokalita": "Lokalita", "diagnoza": "Diagnoza"})
df2["Lokalita"] = df2["Lokalita"].replace(
    {
        "rektum ": "rektum",
        "rektun": "rektum",
        "retum": "rektum",
        "cekum ": "cekum",
        "transverzum ": "transverzum",
    }
)
df2 = df2.fillna("rektum")
df2
```

```python
df = pd.concat([df1, df2], axis=0)
df
```
