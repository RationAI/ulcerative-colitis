```python
from pathlib import Path

import pandas as pd


DATA_DIR = Path("/mnt/data/FTN/colon/IBD_AI")
PROJECT_DIR = Path("/mnt/projects/inflammatory_bowel_dissease/ulcerative_colitis")
```

```python
df1 = pd.read_excel(DATA_DIR / "IBD_AI_FTN.xlsx", index_col=0)
df1
```

```python
df2 = pd.read_excel(DATA_DIR / "IBD_AI_FTN_doplnek.xlsx", index_col=0)
df2 = df2.dropna()
df2["Nancy"] = df2["Nancy"].astype(int)
df2
```

```python
df = pd.concat([df1, df2], axis=0)
df.index = df.index.str.strip()
df
```

```python
for slide_id in df.index:
    slide_path = DATA_DIR / f"{slide_id.replace('/', '_')}.czi"
    if not slide_path.exists():
        print(f"Missing slide: {slide_path}")
```

```python
for slide in DATA_DIR.glob("*.czi"):
    slide_id = slide.stem.replace("_", "/")
    if slide_id not in df.index:
        print(f"Extra slide: {slide}")
```

```python
df.to_csv(PROJECT_DIR / "IBD_AI.csv", index=True)
```
