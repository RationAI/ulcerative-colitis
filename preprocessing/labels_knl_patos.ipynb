```python
from pathlib import Path

import pandas as pd


DIR = Path("/mnt/data/KNL_PATOS/colon/IBD_AI")
```

```python
df1 = pd.read_excel(DIR / "IBD_AI_Liberec.xlsx", index_col=0)
df1.head()
```

```python
df2 = pd.read_excel(DIR / "IBD_AI_Liberec_02.xlsx", index_col=0)
df2.head()
```

```python
df3 = pd.read_excel(DIR / "IBD_AI_Liberec_10_2025.xlsx", index_col=0)
df3.head()
```

```python
df4 = pd.read_excel(DIR / "IBD_AI_Liberec_28_10_2025.xlsx", index_col=0)
df4.head()
```

```python
df = pd.concat([df1, df2, df3, df4], axis=0)
df
```

```python
for name in df.index:
    if not (DIR / name).with_suffix(".czi").exists():
        print(f"Missing file for {name}")
```

```python
for file in DIR.glob("*.czi"):
    name = file.stem
    if name not in df.index:
        print(f"Extra file: {name}")
```

```python
df.to_csv(DIR / "IBD_AI.csv", index=True)
```
